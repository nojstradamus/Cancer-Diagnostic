```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['Unnamed: 32'].isnull().sum()
```

```python
df = df.drop(['Unnamed: 32', 'id'], 1)
```

```python
df.head()
```

```python
X = df.drop(['diagnosis'], 1)
y = df.diagnosis
```

```python
for i in range(len(y)):
    if y[i] == 'M':
        y[i] = 1
    else:
        y[i] = 0
```

```python
y.unique()
```

```python
sns.set_style('darkgrid')
rcParams['figure.figsize'] = 10,7

for i in X.columns:
    print(i)
    print(X[i].describe())
    plt.hist(X[i])
    plt.show()
    print()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
corrmat = X.corr()
```

```python
sns.set_style('darkgrid')
rcParams['figure.figsize'] = 10,7

sns.heatmap(corrmat)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X, y)
```

```python
X
```

```python
y
```

```python
y = y.astype(int)
```

```python
logreg.fit(X, y)
```

```python
logreg.score(X, y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(logreg, X, y, cv= 10)
print(score)
print()
print(score.mean())
```

```python
knn.fit(X, y)
```

```python
knn.score(X, y )
```

```python
score = cross_val_score(knn, X, y, cv=10)
print(score)
print()
print(score.mean())
```

```python
from sklearn import ensemble
```

```python
rfc = ensemble.RandomForestClassifier()
rfc.fit(X, y)
rfc.score(X, y)
```

```python
score = cross_val_score(rfc, X, y, cv= 10)
print(score)
print()
print(score.mean())
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='linear')
```

```python
svm.fit(X, y)
```

```python
svm.score(X, y)
```
